```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6,7]
y = [10,20,30,40,50,60,70]

print(x)
print(y)
```

```python
plt.plot(x,y)
#colocar labels para os eixos + dar titulo

plt.xlabel("x-axis label")
plt.ylabel("y-axis label")
plt.title("my 1st plot")

plt.show(x,y)

```

```python
# você também pode criar subplots dentro de um plot, para fins de comparação, como abaixo
#neste caso, iremos criar 3 subplots (1 linha, 3 colunas), então dentro de subplot() 
#o primeiro será a linha o segundo a quantidade de subplots e o terceiro a coluna

plt.subplot(131)

plt.title("1st plot")

plt.plot(x,y)


#agora a segunda coluna


plt.subplot(132) #observe a troca do número

plt.title("2nd plot")

plt.plot(y,x)

#plt.show(y,x)

#e a terceira

plt.subplot(133)

plt.title("3rd plot")

plt.plot(x,x)

#lembre-se de chamar o show apenas no fim de tudo sem argumentos

plt.show()
```

```python
#importar alguns casos para fazer visualização em boxplot:

market_fact = pd.read_csv("market_fact.csv")

cust_dimen = pd.read_csv("cust_dimen.csv")

prod_dimen = pd.read_csv("prod_dimen.csv")

shipping_dimen = pd.read_csv("shipping_dimen.csv")

orders_dimen = pd.read_csv("orders_dimen.csv")

#visualizar as primeiras 5 linhas

market_fact.head()
```

```python
plt.boxplot(market_fact["Order_Quantity"])

plt.show()
```

```python
#fazer o boxplot das vendas
plt.boxplot(market_fact["Sales"])
plt.title("Primeiro teste")
plt.show()

#alinhar utilizando o log para melhor visualização das vendas e dos outliers
plt.boxplot(market_fact["Sales"])
plt.yscale("log")
plt.title("Segundo teste")
plt.show()
```

```python
#agora iremos criar um histograma com a frequência de determinados valores de vendas

plt.hist(market_fact["Sales"])

plt.xlabel("Sales") #nomeia o eixo x/horizontal como vendas
plt.ylabel("Frequence") #nomeia o eixo y/vertical como a frequencia
plt.title("Sales Frequence Histogram")

plt.show()

#a partir dai podemos observar que quanto maior o valor da compra, menor a frequência
```

```python
#agora iremos fazer um scatter plot para determinar a relação
# das variáveis sales x profit

plt.scatter(market_fact["Sales"], market_fact["Profit"])
plt.xlabel("Sales") 
plt.ylabel("Profit")
plt.title("Relation between Sales and Profit")
plt.show()
```

```python
#hora de usar o seaborn

import seaborn as sns

```

```python
#para customizar a aparencia do seaborn, cheque sua documentação e use:

sns.set_style("darkgrid")

#para fazer um histograma/plot de densidade usaremos a função distplot:

sns.distplot(market_fact["Shipping_Cost"])
plt.show()

#para que o gráfico tenha barras mais visíveis, deve-se utilizar um arg rug na função:

sns.distplot(market_fact["Shipping_Cost"], rug = True)
plt.show()
```

```python
#para que seja apenas tenha a densidade, você usará o argumento hist:

sns.distplot(market_fact["Shipping_Cost"], hist = False)
plt.show()
```

```python
#o default do seaborn em boxplots é no eixo x, como pode ser visto abaixo:

sns.boxplot(market_fact["Order_Quantity"])
plt.show()

#segue abaixo como colocá-lo no eixo y:

sns.boxplot(y=market_fact["Order_Quantity"])
plt.show()
```

```python
#podemos usar a função jointplot para fazer uma distribuição bi-variada

sns.jointplot("Sales", "Profit", market_fact)
plt.show()
```

```python
#além disso podemos criar uma matrix de correlação de todas as variáveis
#no caso colunas no dataset que temos utilizando a função pairplot

sns.pairplot(market_fact)
plt.show()
plt.savefig("corrmatrix")
```

```python
#para as próximas análises, iremos combinar alguns dataframes
mark_prod = pd.merge(market_fact, prod_dimen, how = "inner", on="Prod_id")
mark_prod.head()
```

```python
sns.boxplot(x="Product_Category", y="Sales", data = mark_prod)
plt.xlabel("Product Category")
plt.yscale("log") #como fizemos o boxplot antes do sales 
plt.show()
```

```python
sns.barplot(x="Product_Category", y="Sales", data=mark_prod) # os valores são agregados por média aq
plt.xlabel("Product Category")
plt.title("Mean Sales for Product Categories")
plt.show()
```

```python
#podemos adicionar o argumento hue para colocar mais categorias
#criaremos um novo df agora com dados dos consumidores
mark_prod_cust = pd.merge(mark_prod, cust_dimen, how = "inner", on = "Cust_id")

#usaremos tb o arg estimador
sns.barplot(x="Product_Category", y="Sales", hue = "Customer_Segment", data=mark_prod_cust, estimator = np.mean)
plt.xlabel("Product Category")
plt.title("Mean Sales for Product Category By Segment")
plt.show()
```

```python
#agora iremos fazer a mesma coisa mas usando como estimador a soma das vendas

sns.barplot(x="Product_Category", y="Sales", hue = "Customer_Segment", data=mark_prod_cust, estimator = np.sum)
plt.xlabel("Product Category")
plt.title("Most Sales for Product Category By Segment")
plt.show()
```

```python
#para plotarmos em relação a subcategorias, iremos colocar 
#as subcategorias no eixo y pela legibilidade

plt.figure(figsize=(15,8)) #para aumentar o tamanho do grafico
sns.barplot(y="Product_Sub_Category", x="Sales", data=mark_prod_cust, estimator=np.sum)
plt.ylabel("Product Sub Category")
plt.title("Sum of Sales by Subcategory")
plt.savefig("barplot2")
plt.show()
```

```python
#para explorar outros tipos de plot, iremos mergir o market_fact com o order_dimen

mark_ord = pd.merge(market_fact, orders_dimen, how="inner", on="Ord_id")

mark_ord.head()

#como este df tem datas, iremos usar a função to_datetime do pandas
mark_ord["Order_Date"] = pd.to_datetime(mark_ord["Order_Date"])
```

```python
#após isso iremos agrupar as vendas pela data de pedido com o groupby

date_sales = mark_ord.groupby("Order_Date")["Sales"].sum()
print(date_sales.head())

print(date_sales)
```

```python
#agora para plotar o tempo, usaremos o sns e sua função lineplot

plt.figure(figsize=(22,12))

sns.lineplot(data = date_sales)
plt.show()
```

```python
#caso vc queira extrair alguma coisa das datas parar criar uma nova coluna
#vc pode usar o .dt
#extrair mês
mark_ord["Month"] = mark_ord["Order_Date"].dt.month
#extrair ano
mark_ord["Year"] = mark_ord["Order_Date"].dt.year

mark_ord.head()
```

```python
#agora podemos criar uma nova variável para analisar

year_mon = pd.DataFrame(mark_ord.groupby(["Year", "Month"]).Sales.mean()) #média de vendas por mês do ano

year_mon.head()
```

```python
year_mon2 = pd.pivot_table(mark_ord, values = "Sales", index = "Year", columns = "Month", aggfunc= "mean")
year_mon2
```

```python
plt.figure(figsize=(12,8))

sns.heatmap(year_mon, cmap= "YlGnBu")

plt.show()
```
