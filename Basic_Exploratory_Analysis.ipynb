```python
cust_dimen = pd.read_csv("cust_dimen.csv")

prod_dimen = pd.read_csv("prod_dimen.csv")

shipping_dimen = pd.read_csv("shipping_dimen.csv")

orders_dimen = pd.read_csv("orders_dimen.csv")
```

```python
market_cust = pd.merge(marketfact_df, cust_dimen, how = "inner", on = "Cust_id")
market_cust.head()
```

```python
market_cust_pr = pd.merge(market_cust, prod_dimen, how = "inner", on = "Prod_id")
market_cust_pr.head()
```

```python
market_cust_pr_ship = pd.merge(market_cust_pr, shipping_dimen, how = "inner", on = "Ship_id")
market_cust_pr_ship_ord = pd.merge(market_cust_pr_ship, orders_dimen, how = "inner", on = "Ord_id")
market_cust_pr_ship_ord.head()
```

```python
group_prod_cat = market_cust_pr_ship_ord.groupby(["Product_Category", "Product_Sub_Category"])
group_prod_cat

pd.DataFrame(group_prod_cat.Sales.sum())
```

```python

pd.DataFrame(group_prod_cat.Sales.sum()/sum(market_cust_pr_ship_ord.Sales)*100)
```

```python
market_cust_pr_ship_ord.pivot_table(values = "Sales", index = "Product_Category", aggfunc = "mean")
```

```python
market_cust_pr_ship_ord.pivot_table(values = "Sales", index = "Product_Category", aggfunc = "median")
```

```python
market_cust_pr_ship_ord.pivot_table(values = "Sales", index = "Product_Category", aggfunc = "max")
```
